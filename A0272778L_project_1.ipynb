```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np
import numpy.random as npr
import math
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
style = {'description_width': 'initial'}
S0 = widgets.IntText(value=1.0, description='Initial Stock Price (S0):', style=style)
r = widgets.FloatText(value=0.0, description='Short rate (r):', style=style)
sigma = widgets.FloatText(value=0.0, min=0.0, max=1.0, description='Volatility:', style=style)
start_date = widgets.DatePicker(description='Select Start Date:', value=datetime.date.today(),disabled=False)
end_date = widgets.DatePicker(description='Select End Date:', value = datetime.date.today(), disabled=False)
freq = widgets.Dropdown(options=['W','M','Y'], value='W', description='Frequency', disabled=False)
I = widgets.IntText(value=1.0, description='Number of Simulations', style=style)
```

```python
def gbm(S0, r, sigma, I, start_date, end_date, freq):
    time_grid = pd.date_range(start = start_date, end=end_date, freq = freq).to_pydatetime()
    M = len(time_grid)
    S = np.zeros((M, I)) 
    S[0] = S0
    for t in range(1, M):
        dt = (time_grid[t] - time_grid[t-1]).days/365
        S[t] = S[t-1] * np.exp((r - 0.5*sigma**2)*dt + sigma*math.sqrt(dt) * npr.standard_normal(I))          
    return S
```

```python
output_area = widgets.Output()
def stock_mean(b):
    S = gbm(S0.value, r.value, sigma.value, I.value, start_date.value, end_date.value, freq.value)
    with output_area:
        if sigma.value < 0.0 or sigma.value > 1.0:
            print("ERROR: Volatility must be between 0.0 and 1.0.")
            return
        if end_date.value <= start_date.value:
            print("ERROR: End Date must be after Start Date.")
            return
        clear_output(wait=True)
        avg = S[-1].mean()
        print(f"📊 The Mean Stock Price is: ${avg:.2f}")

def plot_stock(b):
    S = gbm(S0.value, r.value, sigma.value, I.value, start_date.value, end_date.value, freq.value)
    with output_area:
        if sigma.value < 0.0 or sigma.value > 1.0:
            print("ERROR: Volatility must be between 0.0 and 1.0.")
            return
        if end_date.value <= start_date.value:
            print("ERROR: End Date must be after Start Date.")
            return
        clear_output(wait=True)
        plt.figure(figsize=(6, 3))
        plt.plot(S[:,:10], lw=1.5)
        plt.title("First 10 Simulated Paths")
        plt.ylabel("Price ($)")
        plt.xlabel("Time step")
        plt.show()

button_mean = widgets.Button(description="Calculate Mean",button_style='info')
button_plot = widgets.Button(description="Generate Plot", button_style='success')

button_mean.on_click(stock_mean)
button_plot.on_click(plot_stock)

input_widgets = widgets.VBox([S0, r, sigma, I, start_date, end_date, freq])
buttons = widgets.HBox([button_mean, button_plot])
layout = widgets.VBox([input_widgets, buttons, output_area])

display(layout)
```
